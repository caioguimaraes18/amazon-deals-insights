```python
import pandas as pd
import json


with open('../Dataset/amazon_deals.json', 'r') as f:
    data = json.load(f)
    
df = pd.json_normalize(data['data']['deals'], sep='.')

df.head()

```

```python
df_dim = df[[
    'product_asin',
    'deal_type',
    'deal_title',
    'deal_state',
    'deal_url',
    'deal_photo',    
    
]]

df_dim.head(10)
```

```python
df_fact = df[[
    'deal_id',
    'product_asin',
    'deal_price.amount',
    'list_price.amount',
    'savings_amount.amount',
    'deal_price.currency',
    'savings_percentage',
    'deal_starts_at',
    'deal_ends_at'
]]

df_fact.head(10)
```

```python
df_fact = df_fact.rename(columns={
    'deal_price.amount': 'deal_price',
    'list_price.amount': 'list_price',
    'savings_amount.amount': 'savings_amount',
    'deal_price.currency': 'deal_currency',
})

num_cols = ['deal_price', 'list_price', 'savings_amount', 'savings_percentage']
time_cols = ['deal_starts_at', 'deal_ends_at']

for col in num_cols:
    df_fact[col] = pd.to_numeric(df_fact[col])
    
for col in time_cols:
    df_fact[col] = pd.to_datetime(df_fact[col]).dt.tz_convert('America/Sao_Paulo')
    

df_fact
```

```python
# Verificando duplicatas

print(f'O total de duplicatas na dimensão é: {df_dim.duplicated().sum()}')
print(f'O total de duplicatas no fato é: {df_fact.duplicated().sum()}')

# Contagem de linhas

print(f"Dimensão: {df_dim.shape[0]} linhas, {df_dim.shape[1]} colunas")
print(f"Fato: {df_fact.shape[0]} linhas, {df_fact.shape[1]} colunas")
```

```python
df_final = df_dim.merge(df_fact, on ='product_asin', how='left')

df_final
```

```python
import matplotlib.pyplot as plt
import textwrap

# Pega os 10 produtos com maior % de desconto
top_10 = df_final.sort_values(by='savings_percentage', ascending=False).head(10).copy()

# Limita os títulos a no máximo 2 linhas de 30 caracteres
def resumir_titulo(titulo):
    linhas = textwrap.wrap(titulo, width=30)
    return '\n'.join(linhas[:2]) + ('...' if len(linhas) > 2 else '')

top_10['titulo_resumido'] = top_10['deal_title'].apply(resumir_titulo)

# Inverter ordem para o maior no topo
top_10 = top_10[::-1]

# Tamanho do gráfico
plt.figure(figsize=(10, 6))

# Gráfico horizontal
barras = plt.barh(top_10['titulo_resumido'], top_10['savings_percentage'], color='#FF6709', label='Desconto (%)')

# Valor ao lado da barra
for barra in barras:
    xval = barra.get_width()
    plt.text(xval + 0.5, barra.get_y() + barra.get_height()/2,
             f'{xval:.1f}%', va='center', fontsize=9)

# Títulos e eixos
plt.title('Top 10 produtos com maior percentual de desconto', fontsize=14)
plt.xlabel('Porcentagem de Desconto (%)', fontsize=12)
plt.ylabel('Produtos', fontsize=12)

# Legenda e layout
plt.legend()
plt.tight_layout()
plt.show()
```
